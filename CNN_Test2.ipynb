```
size = 600
training_path = 'gdrive/My Drive/AppliedProject/EasyData/Training'
test_path = 'gdrive/My Drive/AppliedProject/EasyData/Test'


cnn_layers = [Conv2D(64,(3,3),padding='same',activation='relu',input_shape=(size,size,1)), MaxPool2D(),
              Conv2D(128,(3,3),padding='same',activation='relu'), MaxPool2D(),
              Conv2D(128,(3,3),padding='same',activation='relu'), MaxPool2D(),
              Conv2D(128,(3,3),padding='same',activation='relu'), MaxPool2D(),
              Conv2D(128,(3,3),padding='same',activation='relu'), MaxPool2D(),
              Conv2D(128,(3,3),padding='same',activation='relu'), MaxPool2D(),
              Conv2D(128,(3,3),padding='same',activation='relu'), MaxPool2D(),
              Conv2D(128,(3,3),padding='same',activation='relu'), MaxPool2D(),
              Conv2D(128,(3,3),padding='same',activation='relu'), MaxPool2D(),
              Flatten(),
              Dense(128,activation='relu'),
              Dense(64,activation='relu'),
              Dense(1,activation='sigmoid')]

classifier = keras.Sequential(cnn_layers)

classifier.compile(optimizer=keras.optimizers.Adam(lr=1e-6), loss='binary_crossentropy', metrics=['accuracy'])
```

```
train_datagen = ImageDataGenerator(rescale = 1./255,
shear_range = 0.2,
zoom_range = 0.2,
horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)

training_set = train_datagen.flow_from_directory(training_path,
target_size = (256,256),
batch_size = 16,
class_mode = 'binary',
shuffle=True,
color_mode='grayscale')

print(training_set.labels)

test_set = test_datagen.flow_from_directory(test_path,
target_size = (256,256),
batch_size = 16,
class_mode = 'binary',
shuffle=False,
color_mode = 'grayscale',)
```

```
classifier.fit_generator(training_set,
steps_per_epoch = 400 // 16,
epochs = 50,
validation_data = test_set,
validation_steps = (80 // 16))
```
